```python
# â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
# â•‘  CELDA 2: CÃ“DIGO PRINCIPAL DEL EVALUADOR                         â•‘
# â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

"""
EVALUADOR AUTOMATIZADO DE PRUEBAS DE CONTABILIDAD
VersiÃ³n 3.0 - Google Colab Edition
Sistema GrupoCAP
Autor: MoisesElvir
Fecha: 2025-11-21
"""

import openpyxl
import json
import io
from datetime import datetime
from google.colab import files
import matplotlib.pyplot as plt
import numpy as np

class EvaluadorContabilidad:
    """Clase principal para evaluar pruebas de contabilidad."""

    def __init__(self, archivo_bytes):
        """Inicializa el evaluador con el archivo subido."""
        self.archivo_bytes = archivo_bytes
        self.wb = None
        self.resultados = {}

    def cargar_archivo(self):
        """Carga el archivo de Excel."""
        try:
            self.wb = openpyxl.load_workbook(self.archivo_bytes)
            print("âœ… Archivo cargado exitosamente")
            print(f"ğŸ“Š Hojas disponibles: {', '.join(self.wb.sheetnames)}")
            return True
        except Exception as e:
            print(f"âŒ Error al cargar el archivo: {e}")
            return False

    def evaluar_prueba_costos(self):
        """EvalÃºa la Prueba de AnÃ¡lisis de Costos y Presupuestos."""
        print("\n" + "=" * 70)
        print("ğŸ“ˆ EVALUANDO: PRUEBA DE ANÃLISIS DE COSTOS Y PRESUPUESTOS")
        print("=" * 70)

        resultados = {
            "nombre_prueba": "AnÃ¡lisis de Costos y Presupuestos",
            "puntaje_total": 50,
            "puntaje_obtenido": 0,
            "secciones": {},
            "timestamp": datetime.now().isoformat(),
            "usuario": "MoisesElvir"
        }

        # Evaluar cada secciÃ³n
        if "COSTOS_POR_PROCESOS" in self.wb.sheetnames:
            print("\nğŸ” Evaluando: Costos por Procesos...")
            ws = self.wb["COSTOS_POR_PROCESOS"]
            seccion = self._evaluar_costos_procesos(ws)
            resultados["secciones"]["costos_procesos"] = seccion
            resultados["puntaje_obtenido"] += seccion["puntaje"]
        else:
            print("âš ï¸  Hoja 'COSTOS_POR_PROCESOS' no encontrada")

        if "PUNTO_EQUILIBRIO" in self.wb.sheetnames:
            print("ğŸ” Evaluando: Punto de Equilibrio...")
            ws = self.wb["PUNTO_EQUILIBRIO"]
            seccion = self._evaluar_punto_equilibrio(ws)
            resultados["secciones"]["punto_equilibrio"] = seccion
            resultados["puntaje_obtenido"] += seccion["puntaje"]
        else:
            print("âš ï¸  Hoja 'PUNTO_EQUILIBRIO' no encontrada")

        if "PRESUPUESTO_EFECTIVO" in self.wb.sheetnames:
            print("ğŸ” Evaluando: Presupuesto de Efectivo...")
            ws = self.wb["PRESUPUESTO_EFECTIVO"]
            seccion = self._evaluar_presupuesto_efectivo(ws)
            resultados["secciones"]["presupuesto_efectivo"] = seccion
            resultados["puntaje_obtenido"] += seccion["puntaje"]
        else:
            print("âš ï¸  Hoja 'PRESUPUESTO_EFECTIVO' no encontrada")

        print("\nâœ… EvaluaciÃ³n completada")
        return resultados

    def _evaluar_costos_procesos(self, ws):
        """EvalÃºa costos por procesos."""
        puntaje = 0
        max_puntaje = 15
        feedback = []

        respuestas_correctas = {
            'unidades_equiv': 9000,
            'costo_unitario': 100,
            'costo_terminadas': 800000,
            'costo_proceso': 100000
        }

        evaluaciones = [
            ('B22', 'unidades_equiv', 4, 'Unidades equivalentes'),
            ('B23', 'costo_unitario', 4, 'Costo por unidad equivalente'),
            ('B24', 'costo_terminadas', 4, 'Costo de unidades terminadas'),
            ('B25', 'costo_proceso', 3, 'Costo de inventario en proceso')
        ]

        try:
            for celda, clave, puntos, descripcion in evaluaciones:
                valor = ws[celda].value
                correcto = respuestas_correctas[clave]
                tolerancia = 100 if 'costo' in clave else 1

                if valor and abs(float(valor) - correcto) < tolerancia:
                    puntaje += puntos
                    feedback.append(f"âœ… {descripcion}: CORRECTO")
                else:
                    feedback.append(f"âŒ {descripcion}: INCORRECTO (Correcto: {correcto:,.0f})")
        except Exception as e:
            feedback.append(f"âŒ Error: {e}")

        return {
            "nombre": "Costos por Procesos",
            "puntaje": puntaje,
            "max_puntaje": max_puntaje,
            "feedback": feedback
        }

    def _evaluar_punto_equilibrio(self, ws):
        """EvalÃºa punto de equilibrio."""
        puntaje = 0
        max_puntaje = 15
        feedback = []

        respuestas_correctas = {
            'pe_unidades': 1000,
            'pe_lempiras': 500000,
            'unidades_utilidad': 1500
        }

        evaluaciones = [
            ('B13', 'pe_unidades', 5, 'Punto de equilibrio en unidades'),
            ('B14', 'pe_lempiras', 5, 'Punto de equilibrio en lempiras'),
            ('B15', 'unidades_utilidad', 5, 'Unidades para utilidad objetivo')
        ]

        try:
            for celda, clave, puntos, descripcion in evaluaciones:
                valor = ws[celda].value
                correcto = respuestas_correctas[clave]
                tolerancia = 100 if 'lempiras' in clave else 1

                if valor and abs(float(valor) - correcto) < tolerancia:
                    puntaje += puntos
                    feedback.append(f"âœ… {descripcion}: CORRECTO")
                else:
                    feedback.append(f"âŒ {descripcion}: INCORRECTO (Correcto: {correcto:,.0f})")
        except Exception as e:
            feedback.append(f"âŒ Error: {e}")

        return {
            "nombre": "Punto de Equilibrio",
            "puntaje": puntaje,
            "max_puntaje": max_puntaje,
            "feedback": feedback
        }

    def _evaluar_presupuesto_efectivo(self, ws):
        """EvalÃºa presupuesto de efectivo."""
        puntaje = 0
        max_puntaje = 20
        feedback = []

        feedback.append("âš ï¸  Esta secciÃ³n requiere revisiÃ³n manual")

        if ws['B15'].value is not None:
            puntaje += 10
            feedback.append("âœ… SecciÃ³n completada (puntaje parcial: 10/20)")
        else:
            feedback.append("âŒ SecciÃ³n incompleta")

        return {
            "nombre": "Presupuesto de Efectivo",
            "puntaje": puntaje,
            "max_puntaje": max_puntaje,
            "feedback": feedback
        }

    def evaluar_prueba_impuestos(self):
        """EvalÃºa la Prueba de Impuestos y NÃ³mina."""
        print("\n" + "=" * 70)
        print("ğŸ’° EVALUANDO: PRUEBA DE IMPUESTOS Y NÃ“MINA")
        print("=" * 70)

        resultados = {
            "nombre_prueba": "Impuestos y NÃ³mina",
            "puntaje_total": 50,
            "puntaje_obtenido": 0,
            "secciones": {},
            "timestamp": datetime.now().isoformat(),
            "usuario": "MoisesElvir"
        }

        if "CALCULO_ISV" in self.wb.sheetnames:
            print("\nğŸ” Evaluando: CÃ¡lculo de ISV...")
            ws = self.wb["CALCULO_ISV"]
            seccion = self._evaluar_isv(ws)
            resultados["secciones"]["isv"] = seccion
            resultados["puntaje_obtenido"] += seccion["puntaje"]
        else:
            print("âš ï¸  Hoja 'CALCULO_ISV' no encontrada")

        if "CALCULO_NOMINA" in self.wb.sheetnames:
            print("ğŸ” Evaluando: CÃ¡lculo de NÃ³mina...")
            ws = self.wb["CALCULO_NOMINA"]
            seccion = self._evaluar_nomina(ws)
            resultados["secciones"]["nomina"] = seccion
            resultados["puntaje_obtenido"] += seccion["puntaje"]
        else:
            print("âš ï¸  Hoja 'CALCULO_NOMINA' no encontrada")

        if "RETENCIONES_ISR" in self.wb.sheetnames:
            print("ğŸ” Evaluando: Retenciones ISR...")
            ws = self.wb["RETENCIONES_ISR"]
            seccion = self._evaluar_retenciones(ws)
            resultados["secciones"]["retenciones"] = seccion
            resultados["puntaje_obtenido"] += seccion["puntaje"]
        else:
            print("âš ï¸  Hoja 'RETENCIONES_ISR' no encontrada")

        print("\nâœ… EvaluaciÃ³n completada")
        return resultados

    def _evaluar_isv(self, ws):
        """EvalÃºa ISV."""
        puntaje = 0
        max_puntaje = 15
        feedback = []
        isv_correcto = 12000

        try:
            isv = ws['B13'].value
            if isv and abs(float(isv) - isv_correcto) < 10:
                puntaje += 15
                feedback.append("âœ… ISV por pagar: CORRECTO")
            else:
                feedback.append(f"âŒ ISV por pagar: INCORRECTO (Correcto: L {isv_correcto:,.2f})")
                feedback.append("   FÃ³rmula: (Ventas Ã— 15%) - (Compras Ã— 15%)")
        except Exception as e:
            feedback.append(f"âŒ Error: {e}")

        return {"nombre": "CÃ¡lculo de ISV", "puntaje": puntaje, "max_puntaje": max_puntaje, "feedback": feedback}

    def _evaluar_nomina(self, ws):
        """EvalÃºa nÃ³mina."""
        puntaje = 0
        max_puntaje = 20
        feedback = []

        empleados = [
            {"nombre": "Juan PÃ©rez", "ihss": 1050, "rap": 225, "neto": 13725},
            {"nombre": "MarÃ­a LÃ³pez", "ihss": 1260, "rap": 270, "neto": 16470},
            {"nombre": "Carlos GÃ³mez", "ihss": 840, "rap": 180, "neto": 10980},
            {"nombre": "Ana MartÃ­nez", "ihss": 1400, "rap": 300, "neto": 18300}
        ]

        try:
            correctos = 0
            for i, emp in enumerate(empleados, 7):
                ihss = ws.cell(row=i, column=3).value
                rap = ws.cell(row=i, column=4).value
                neto = ws.cell(row=i, column=5).value

                if ihss and abs(float(ihss) - emp["ihss"]) < 1: correctos += 1
                if rap and abs(float(rap) - emp["rap"]) < 1: correctos += 1
                if neto and abs(float(neto) - emp["neto"]) < 1: correctos += 1

            puntaje = int((correctos / 12) * max_puntaje)
            porcentaje = (correctos / 12) * 100
            feedback.append(f"ğŸ“Š CÃ¡lculos correctos: {correctos}/12 ({porcentaje:.1f}%)")

            if correctos == 12:
                feedback.append("âœ… Todos los cÃ¡lculos son CORRECTOS")
            else:
                feedback.append("âš ï¸  Algunos cÃ¡lculos tienen errores")
                feedback.append("   Recordatorio: IHSS = 7%, RAP = 1.5%")
        except Exception as e:
            feedback.append(f"âŒ Error: {e}")

        return {"nombre": "CÃ¡lculo de NÃ³mina", "puntaje": puntaje, "max_puntaje": max_puntaje, "feedback": feedback}

    def _evaluar_retenciones(self, ws):
        """EvalÃºa retenciones."""
        puntaje = 0
        max_puntaje = 15
        feedback = []
        retenciones_correctas = 10000

        try:
            retenciones = ws['B11'].value
            if retenciones and abs(float(retenciones) - retenciones_correctas) < 10:
                puntaje += 15
                feedback.append("âœ… Total de retenciones: CORRECTO")
            else:
                feedback.append(f"âŒ Total de retenciones: INCORRECTO (Correcto: L {retenciones_correctas:,.2f})")
                feedback.append("   FÃ³rmula: Suma de servicios Ã— 10%")
        except Exception as e:
            feedback.append(f"âŒ Error: {e}")

        return {"nombre": "Retenciones ISR", "puntaje": puntaje, "max_puntaje": max_puntaje, "feedback": feedback}

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# FUNCIONES DE VISUALIZACIÃ“N
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def graficar_resultados(resultados):
    """Crea visualizaciones de los resultados."""
    secciones = []
    puntajes = []
    maximos = []

    for seccion in resultados['secciones'].values():
        secciones.append(seccion['nombre'])
        puntajes.append(seccion['puntaje'])
        maximos.append(seccion['max_puntaje'])

    # Crear figura con 2 subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

    # GRÃFICO 1: Barras comparativas
    x = np.arange(len(secciones))
    width = 0.35

    bars1 = ax1.bar(x - width/2, maximos, width, label='Puntaje MÃ¡ximo',
                     alpha=0.5, color='#3b82f6', edgecolor='#1e40af')
    bars2 = ax1.bar(x + width/2, puntajes, width, label='Puntaje Obtenido',
                     alpha=0.9, color='#10b981', edgecolor='#059669')

    ax1.set_xlabel('Secciones', fontweight='bold', fontsize=12)
    ax1.set_ylabel('Puntaje', fontweight='bold', fontsize=12)
    ax1.set_title('ğŸ“Š ComparaciÃ³n de Puntajes por SecciÃ³n', fontweight='bold', fontsize=14, pad=20)
    ax1.set_xticks(x)
    ax1.set_xticklabels(secciones, rotation=45, ha='right')
    ax1.legend(loc='upper right')
    ax1.grid(axis='y', alpha=0.3)

    # Agregar valores encima de las barras
    for bar in bars2:
        height = bar.get_height()
        ax1.text(bar.get_x() + bar.get_width()/2., height,
                f'{int(height)}', ha='center', va='bottom', fontweight='bold')

    # GRÃFICO 2: GrÃ¡fico de pastel
    porcentaje_obtenido = resultados['puntaje_obtenido']
    porcentaje_perdido = resultados['puntaje_total'] - resultados['puntaje_obtenido']

    colors = ['#10b981', '#ef4444']
    explode = (0.1, 0)

    ax2.pie([porcentaje_obtenido, porcentaje_perdido],
            labels=['Puntaje Obtenido', 'Puntaje Perdido'],
            autopct='%1.1f%%',
            startangle=90,
            colors=colors,
            explode=explode,
            shadow=True,
            textprops={'fontsize': 11, 'fontweight': 'bold'})

    ax2.set_title(f'ğŸ¯ DistribuciÃ³n del Puntaje Total\n{resultados["nombre_prueba"]}',
                  fontweight='bold', fontsize=14, pad=20)

    plt.tight_layout()
    plt.show()

    # Resumen estadÃ­stico
    porcentaje_total = (resultados['puntaje_obtenido'] / resultados['puntaje_total']) * 100
    print("\n" + "=" * 70)
    print("ğŸ“ˆ RESUMEN ESTADÃSTICO")
    print("=" * 70)
    print(f"Puntaje Total: {resultados['puntaje_obtenido']}/{resultados['puntaje_total']}")
    print(f"Porcentaje: {porcentaje_total:.2f}%")
    print(f"Secciones evaluadas: {len(resultados['secciones'])}")
    print(f"Promedio por secciÃ³n: {resultados['puntaje_obtenido']/len(resultados['secciones']):.2f}")

    if porcentaje_total >= 90:
        clasificacion = "EXCELENTE ğŸŒŸ"
    elif porcentaje_total >= 80:
        clasificacion = "MUY BUENO âœ¨"
    elif porcentaje_total >= 70:
        clasificacion = "BUENO ğŸ‘"
    elif porcentaje_total >= 60:
        clasificacion = "APROBADO âœ“"
    else:
        clasificacion = "NECESITA MEJORAR ğŸ“š"

    print(f"ClasificaciÃ³n: {clasificacion}")
    print("=" * 70)

def generar_reporte(resultados):
    """Genera reporte detallado en texto."""
    porcentaje = (resultados['puntaje_obtenido'] / resultados['puntaje_total']) * 100

    if porcentaje >= 90: clasificacion = "EXCELENTE ğŸŒŸ"
    elif porcentaje >= 80: clasificacion = "MUY BUENO âœ¨"
    elif porcentaje >= 70: clasificacion = "BUENO ğŸ‘"
    elif porcentaje >= 60: clasificacion = "APROBADO âœ“"
    else: clasificacion = "NECESITA MEJORAR ğŸ“š"

    reporte = f"""
â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
â•‘              REPORTE DE EVALUACIÃ“N - GRUPOCAP                     â•‘
â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

ğŸ“‹ INFORMACIÃ“N GENERAL
â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”
Prueba:          {resultados['nombre_prueba']}
Usuario:         MoisesElvir
Fecha:           {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}
Sistema:         GrupoCAP Evaluador v3.0

ğŸ¯ RESULTADOS
â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”
Puntaje Total:   {resultados['puntaje_obtenido']}/{resultados['puntaje_total']}
Porcentaje:      {porcentaje:.2f}%
ClasificaciÃ³n:   {clasificacion}

ğŸ“Š DETALLE POR SECCIONES
â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”
"""

    for seccion in resultados['secciones'].values():
        porcentaje_seccion = (seccion['puntaje'] / seccion['max_puntaje']) * 100
        reporte += f"\nğŸ“Œ {seccion['nombre']}\n"
        reporte += f"   Puntaje: {seccion['puntaje']}/{seccion['max_puntaje']} ({porcentaje_seccion:.1f}%)\n"
        reporte += "   Feedback:\n"
        for fb in seccion['feedback']:
            reporte += f"     â€¢ {fb}\n"

    reporte += f"\n{'â•' * 71}\n"
    reporte += "Generado por: Sistema de EvaluaciÃ³n Automatizada GrupoCAP\n"
    reporte += f"{'â•' * 71}\n"

    return reporte

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# FUNCIONES PRINCIPALES
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def subir_y_evaluar(tipo_prueba='costos'):
    """FunciÃ³n principal para subir y evaluar archivo."""
    print("=" * 70)
    print("ğŸš€ EVALUADOR DE CONTABILIDAD - GOOGLE COLAB")
    print("   GrupoCAP - Sistema de EvaluaciÃ³n Automatizada")
    print("=" * 70)
    print(f"\nğŸ“¤ Tipo de prueba: {tipo_prueba.upper()}")
    print("Por favor, sube tu archivo Excel...")
    print("-" * 70)

    # Subir archivo
    uploaded = files.upload()

    if not uploaded:
        print("âŒ No se subiÃ³ ningÃºn archivo")
        return None

    filename = list(uploaded.keys())[0]
    file_bytes = io.BytesIO(uploaded[filename])

    print(f"\nâœ… Archivo recibido: {filename}")
    print("â³ Procesando evaluaciÃ³n...")
    print("-" * 70)

    # Crear evaluador
    evaluador = EvaluadorContabilidad(file_bytes)

    if not evaluador.cargar_archivo():
        return None

    # Evaluar segÃºn tipo
    if tipo_prueba.lower() == 'costos':
        resultados = evaluador.evaluar_prueba_costos()
    elif tipo_prueba.lower() == 'impuestos':
        resultados = evaluador.evaluar_prueba_impuestos()
    else:
        print(f"âŒ Tipo de prueba no vÃ¡lido: '{tipo_prueba}'")
        return None

    # Mostrar reporte
    print(generar_reporte(resultados))

    # Guardar JSON
    nombre_json = f"reporte_{datetime.now().strftime('%Y%m%d_%H%M%S')}.json"
    with open(nombre_json, 'w', encoding='utf-8') as f:
        json.dump(resultados, f, ensure_ascii=False, indent=4)

    print(f"\nâœ… Reporte JSON guardado: {nombre_json}")

    # Descargar reporte
    try:
        files.download(nombre_json)
        print("â¬‡ï¸  Descargando reporte JSON...")
    except:
        print("âš ï¸  No se pudo iniciar descarga automÃ¡tica")

    return resultados

def menu_principal():
    """Muestra el menÃº de instrucciones."""
    print("""
    â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
    â•‘       SISTEMA DE EVALUACIÃ“N DE CONTABILIDAD - COLAB           â•‘
    â•‘              GrupoCAP - VersiÃ³n 3.0 (2025)                     â•‘
    â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

    ğŸ‘¤ Usuario: MoisesElvir
    ğŸ“… Fecha: 2025-11-21

    ğŸ“– INSTRUCCIONES DE USO:

    â”Œâ”€ OPCIÃ“N 1: Evaluar Prueba de Costos â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
    â”‚                                                                â”‚
    â”‚  resultados = subir_y_evaluar('costos')                       â”‚
    â”‚                                                                â”‚
    â”‚  EvalÃºa:                                                       â”‚
    â”‚  â€¢ Costos por Procesos (15 pts)                               â”‚
    â”‚  â€¢ Punto de Equilibrio (15 pts)                               â”‚
    â”‚  â€¢ Presupuesto de Efectivo (20 pts)                           â”‚
    â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜

    â”Œâ”€ OPCIÃ“N 2: Evaluar Prueba de Impuestos â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
    â”‚                                                                â”‚
    â”‚  resultados = subir_y_evaluar('impuestos')                    â”‚
    â”‚                                                                â”‚
    â”‚  EvalÃºa:                                                       â”‚
    â”‚  â€¢ CÃ¡lculo de ISV (15 pts)                                    â”‚
    â”‚  â€¢ CÃ¡lculo de NÃ³mina (20 pts)                                 â”‚
    â”‚  â€¢ Retenciones ISR (15 pts)                                   â”‚
    â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜

    â”Œâ”€ OPCIÃ“N 3: Ver GrÃ¡ficos â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
    â”‚                                                                â”‚
    â”‚  graficar_resultados(resultados)                              â”‚
    â”‚                                                                â”‚
    â”‚  Muestra grÃ¡ficos interactivos con los resultados             â”‚
    â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜

    âš¡ EJEMPLO RÃPIDO:

       # 1. Evaluar prueba de costos
       resultados = subir_y_evaluar('costos')

       # 2. Ver grÃ¡ficos
       graficar_resultados(resultados)

    ğŸ’¡ CARACTERÃSTICAS:
    â€¢ âœ… EvaluaciÃ³n automÃ¡tica en segundos
    â€¢ ğŸ“Š GrÃ¡ficos profesionales con matplotlib
    â€¢ ğŸ“¥ Descarga automÃ¡tica de reporte JSON
    â€¢ ğŸ¯ CalificaciÃ³n precisa con tolerancias

    """)

print("âœ… CÃ³digo del evaluador cargado exitosamente")
print("â”" * 70)
print("ğŸ“š Para ver las instrucciones, ejecuta:")
print("   menu_principal()")
print("â”" * 70)
```
